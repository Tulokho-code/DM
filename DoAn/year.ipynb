## Biến Đổi Các Cột form - period - accepted - version Thành Cột year_q

```python
df = pd.read_csv('submissions.csv')

# Chuyển đổi cột period thành datetime
df['period'] = pd.to_datetime(df['period'])

# Tạo cột year_q từ year và quarter của period
df['year_q'] = df['period'].dt.year.astype(str) + '-Q' + df['period'].dt.quarter.astype(str)

print("Các dòng đầu tiên year_q mới")
display(df[['form', 'period', 'accepted', 'version', 'year_q']].head(10))

print("\nDanh sách các year_q duy nhất")
print(df['year_q'].unique())

print("\nPhân phối theo year_q")
print(df['year_q'].value_counts().sort_index())

# Xóa bỏ các cột được gôm
df = df.drop(columns=['form', 'period', 'accepted', 'version'])

# Tạo file mới
df.to_csv('submissions_with_year_q.csv', index=False)
```

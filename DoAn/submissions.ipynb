```python
import pandas as pd

# Đọc hai file
features = pd.read_csv('features_selected.csv')
submissions = pd.read_csv('submissions.csv')

print(f"Shape features_selected: {features.shape}")
print(f"Shape submissions: {submissions.shape}")

# Kiểm tra cột adsh có tồn tại không
print("Columns in features:", features.columns.tolist())
print("Columns in submissions:", submissions.columns.tolist())

# Merge hai file trên adsh (left join để giữ tất cả features)
merged = pd.merge(
    features, 
    submissions, 
    on='adsh', 
    how='left'          # hoặc 'inner' nếu chỉ muốn các adsh có trong cả hai
)

print(f"✅ Đã merge thành công! Shape merged: {merged.shape}")

# Xem một vài dòng mẫu
print(merged.head())

# Lưu file kết quả
merged.to_csv('features_with_submissions.csv', index=False)
print("💾 Đã lưu file: features_with_submissions.csv")
```

```python
import pandas as pd

# Đọc file submissions
submissions = pd.read_csv('submissions.csv')

print(f"Original shape: {submissions.shape}")
print("Original columns:", submissions.columns.tolist())

# Danh sách cột muốn giữ
selected_columns = [
    'adsh', 
    'cik', 
    'sic', 
    'form', 
    'period', 
    'amendment_adsh', 
    'is_amended'
]

# Kiểm tra xem các cột có tồn tại không
missing_cols = [col for col in selected_columns if col not in submissions.columns]
if missing_cols:
    print(f"⚠️ Thiếu các cột sau: {missing_cols}")
else:
    # Tạo dataframe mới chỉ với các cột cần thiết
    submissions_clean = submissions[selected_columns].copy()
    
    print(f"✅ Đã tạo file mới với shape: {submissions_clean.shape}")
    print(submissions_clean.head())
    
    # Lưu file
    submissions_clean.to_csv('submissions_selected.csv', index=False)
    print("💾 Đã lưu file: submissions_selected.csv")
```

```python
import pandas as pd

# Đọc file
df = pd.read_csv('features_selected.csv')

print(f"Original shape: {df.shape}")
print("Original columns:", df.columns.tolist())

# Các cột muốn giữ
selected_cols = ['adsh', 'tag', 'value']

# Kiểm tra cột có tồn tại không
missing = [col for col in selected_cols if col not in df.columns]
if missing:
    print(f"⚠️ Thiếu cột: {missing}")
else:
    # Tạo dataframe mới
    features_clean = df[selected_cols].copy()
    
    print(f"✅ Đã tạo file mới với shape: {features_clean.shape}")
    print(features_clean.head())
    
    # Lưu file
    features_clean.to_csv('features_adsh_tag_value.csv', index=False)
    print("💾 Đã lưu file: features_adsh_tag_value.csv")
```

```python
import pandas as pd

# Đọc hai file
submissions = pd.read_csv('submissions_selected.csv')
features = pd.read_csv('features_adsh_tag_value.csv')

print(f"Submissions shape: {submissions.shape}")
print(f"Features shape:    {features.shape}")

# Kết hợp hai bảng (mỗi adsh sẽ có nhiều tag/value)
merged = pd.merge(
    features,                    # Bảng chính (features)
    submissions,                 # Bảng bổ sung thông tin submission
    on='adsh',
    how='left'                   # Giữ tất cả các dòng features
)

print(f"✅ Merge hoàn tất! Shape merged: {merged.shape}")
print("\nCác cột sau khi merge:")
print(merged.columns.tolist())

# Xem mẫu 5 dòng đầu
print("\nSample data:")
print(merged.head())

# Sắp xếp cho dễ nhìn
merged = merged.sort_values(by=['adsh', 'tag'])

# Lưu file kết quả
merged.to_csv('features_with_submissions.csv', index=False)
print("\n💾 Đã lưu file: features_with_submissions.csv")
```

```python
# Nếu bạn muốn dạng pivot (mỗi adsh một dòng, mỗi tag một cột)
pivot = merged.pivot_table(
    index=['adsh', 'cik', 'sic', 'form', 'period', 'amendment_adsh', 'is_amended'],
    columns='tag',
    values='value',
    aggfunc='first'          # nếu có trùng thì lấy cái đầu tiên
).reset_index()

pivot.to_csv('features_pivoted.csv', index=False)
print(f"Đã tạo file dạng wide: {pivot.shape}")
```
